```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the Hopfield network
def sigma(x):
    return (1-np.exp(-100*x))/(1+np.exp(-100*x))

def hopfield_network(x, W, b):
    return -0.5 * x + 0.5 * np.dot(W, sigma(x)) + 0.5 * b
```

```python
# W and b
## shape: 0.5*(2x2)x(2xnum) + 0.5*(2x1)
W = np.array([[0, 1], [1, 0]])  # 2x2
b = np.array([0, 0]).reshape(2,1)  # 2x1

# initial conditions
## generate 900 initial state from [-1, 1] x [-1, 1]
x_tmp = []
for i in np.linspace(-1, 1, 30):
    for j in np.linspace(-1, 1, 30):
        x_tmp.append([i, j])
x_initial = np.array(x_tmp).T
```

```python
x_initial.shape
```

```python
# time points: 500
# set h as 0.01
t = 200
h = 0.01

traj_list = []
x = x_initial
for _ in range(t):
    ## x(t+h) = x(t) + h*xt'
    # update 900 states simultaneously
    x = x + hopfield_network(x, W, b)*h

    #print(x.shape)
    #break
    traj_list.append(x)

traj_list = np.array(traj_list)
```

```python
# Plot the phase-plane trajectories
plt.plot(traj_list[:, 0], traj_list[:, 1], alpha=0.8)
plt.title('Trajectories of the Hopfield networks')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
